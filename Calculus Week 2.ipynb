```python
%matplotlib inline

import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import warnings
warnings.filterwarnings('ignore')


# Generate data to plot
x = np.linspace(0, 2 * np.pi, 200)
y = np.exp(np.sin(np.sin(x)))

plt.plot(x, y)
plt.xlim((0, 2 * np.pi))
plt.xlabel(r'$x$')
plt.ylabel(r'$\mathrm{e}^{\sin{x}}$')
```

```python
def f(x):
    return np.sin(x)

t1 = np.arange(-5.0, 5.0, 0.1)

plt.subplot(211)
plt.plot(t1, f(t1))

plt.subplot(212)
plt.plot(t1, np.cos(t1))

```
